## Maximum Tree Depth

Let's first check for the best depth parameter for our decision tree. 
- Create an array for for depth values ranging from 1 - 32. 
- In a loop, train the classifier for each depth value (32 runs) 
- Calculate the training and test AUC for each run 
- Plot a graph to show under/over fitting and optimal value 
- Interpret the results

```python
# Identify the optimal tree depth for given data

depths = list(range(1,33))
train_aucs = []
test_aucs = []

for depth in depths:
    dt = DecisionTreeClassifier(max_depth=depth, criterion = 'entropy')
    dt.fit(X_train, y_train)
    
    y_hat = dt.predict(X_train)
    fpr, tpr, thresholds = roc_curve(y_train, y_hat)
    roc_auc = auc(fpr, tpr)
    train_aucs.append(roc_auc)
    
    y_hat = dt.predict(X_test)
    fpr, tpr, thresholds = roc_curve(y_test, y_hat)
    roc_auc = auc(fpr, tpr)
    test_aucs.append(roc_auc)
    

plt.figure(figsize = (15,9))
plt.plot(depths, train_aucs, label= 'Train AUC', color = 'blue')
plt.plot(depths, test_aucs, label= 'Test AUC', color = 'red')
plt.xlabel('Depth')
plt.ylabel('AUC')
plt.legend()
plt.show()
```

```python
# Identify the optimal min-samples-split for given data


splits = np.linspace(.1,1,10)
train_aucs = []
test_aucs = []

for split in splits:
    dt = DecisionTreeClassifier(criterion = 'entropy', min_samples_split=split)
    dt.fit(X_train, y_train)
    
    y_hat = dt.predict(X_train)
    fpr, tpr, thresholds = roc_curve(y_train, y_hat)
    roc_auc = auc(fpr, tpr)
    train_aucs.append(roc_auc)
    
    y_hat = dt.predict(X_test)
    fpr, tpr, thresholds = roc_curve(y_test, y_hat)
    roc_auc = auc(fpr, tpr)
    test_aucs.append(roc_auc)
    

plt.figure(figsize = (15,9))
plt.plot(splits, train_aucs, label= 'Train AUC', color = 'blue')
plt.plot(splits, test_aucs, label= 'Test AUC', color = 'red')
plt.xlabel('Min Sample Split')
plt.ylabel('AUC')
plt.legend()
plt.show()
```

## Minimum Sample Leafs

Now check for the best `min_samples_leafs` parameter value for our decision tree. 
- Create an array for for `min_samples_leafs` values ranging from 0.1 - 0.5 with an increment of 0.1 
- In a loop, train the classifier for each `min_samples_leafs` value (5 runs) 
- Calculate the training and test AUC for each run 
- Plot a graph to show under/over fitting and optimal value 
- Interpret the results

```python
# Calculate the optimal value for minimum sample leafs

min_sample_leafs = np.linspace(.1,.5,5)
train_aucs = []
test_aucs = []

for min_sample_leaf in min_sample_leafs:
    dt = DecisionTreeClassifier(criterion = 'entropy', min_samples_leaf = min_sample_leaf)
    dt.fit(X_train, y_train)
    
    y_hat = dt.predict(X_train)
    fpr, tpr, thresholds = roc_curve(y_train, y_hat)
    roc_auc = auc(fpr, tpr)
    train_aucs.append(roc_auc)
    
    y_hat = dt.predict(X_test)
    fpr, tpr, thresholds = roc_curve(y_test, y_hat)
    roc_auc = auc(fpr, tpr)
    test_aucs.append(roc_auc)
    

plt.figure(figsize = (15,9))
plt.plot(min_sample_leafs, train_aucs, label= 'Train AUC', color = 'blue')
plt.plot(min_sample_leafs, test_aucs, label= 'Test AUC', color = 'red')
plt.xlabel('min_sample_leafs')
plt.ylabel('AUC')
plt.legend()
plt.show()

# Your code here 
```
